```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plot
from sklearn.naive_bayes import GaussianNB
import math
```

```python
inp = open('train.csv','r')
a = inp.readline()
f = a.split(',')
f = ["".join(i.split()) for i in f]
n = len(f)
train_X = pd.read_csv('train.csv',header=0,names=f)
train_Y = train_X['Survived']
del train_X['Survived']
del train_X['Name']
del train_X['Ticket']
del train_X['Cabin']
del train_X['Embarked']
train_X.head()
inp = open('test.csv','r')
f = f.remove('Survived')
test_X = pd.read_csv('test.csv',header=0,names=f)
del test_X['Name']
del test_X['Ticket']
del test_X['Cabin']
del test_X['Embarked']
# print(train_X)
print(test_X)
```

```python
X = list(train_X.values)
Y = list(train_Y.values)
for i in X:
    if i[2] == "male":
        i[2] = 0
    else:
        i[2] = 1
    if math.isnan(i[3]):
        i[3] = 101.0
        
test = list(test_X.values)
for i in test:
    if i[2] == "male":
        i[2] = 0
    else:
        i[2] = 1
    if math.isnan(i[3]):
        i[3] = 101.0
    if math.isnan(i[6]):
        i[6] = 0.0
print(test[152])        
```

```python
clf = GaussianNB()
clf.fit(X,Y)
t_Y = clf.predict(test)
out = open('gender_submission.csv','w')
count = 0
for i in t_Y:
    s = str(test[count][0])+","+str(i)
    print(test[count][0])
    count = count+1
    out.write(s)
    out.write("\n")
# print(t_Y)
# print(Y)
```
